```python
import pandas as pd
import numpy as np


df = pd.read_csv('../media/FinancialJournal.csv', sep=",", header=0)
df['timestamp'] = pd.to_datetime(df.timestamp)
df.sort_values(by=['timestamp','participantId'],ascending=True, inplace=True)
data = df.to_numpy()
```

```python

index = {}

header = []
idx = 0
```

```python
sameTimestamp = []
time = data[0][2].date()

aux = []
index['Wage'] = 0 #necesitamos un valor para tener el max
keys = []
keys.append('Wage')  #el primero es cero sino no
for row in data:
    if row[3] not in index:
        keys.append(row[3]) #esto es para printear despues
        index[row[3]] = index[max(index, key=index.get)] + 1 #le da el index siguiente al pId
    if row[2].date() != time: #si es un dia despues agregamos uno mas
        sameTimestamp.append(aux)
        time = row[2].date()
        aux = []
    aux.append(row)
print(len(sameTimestamp))
spendings = np.zeros((df['category'].nunique(), len(sameTimestamp))) #array que guarda los gastos por dia
# for id in index:
#     spendings[index[id]][0] = int(id)
```

```python
import csv
from io import StringIO

for sT in sameTimestamp:
    #print(idx)
    header.append(sT[0][2].date()) #agrega la fecha al header
    if idx != 0:
        spendings[:,idx] = spendings[:, idx - 1]
    for row in sT:
        spendings[index[row[3]]][idx] = spendings[index[row[3]]][idx - 1] + row[4]
    idx += 1


with open('/Users/franciscochoi/Desktop/infovis/tpVast/res/ej12.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    print(len(keys))
    print(index)
    header.insert(0, 'ParticipantId')
    writer.writerow(header)
    # write multiple rows
    it = 0
    for row in spendings:
        print(it)
        s = StringIO()
        np.savetxt(s, row, fmt='%.5f', newline=",")
        #VIstring = ','.join(['%.5f' % num for num in row])
        print(s.getvalue())
        f.write(keys[it] + "," + s.getvalue() + "\n")
        it += 1

    f.close()

print("Finish")
```
