```python
import pandas as pd


finJ = pd.read_csv('../media/FinancialJournal.csv', sep=",", header=0)
part = pd.read_csv('../media/Participants.csv', sep=",", header=0)
df = pd.merge(part, finJ, on="participantId", how="inner")
```

```python
data = df.to_numpy()
cats = df.category.unique()

res = {}
max_x = df.loc[df['age'].idxmin()]

for i in range(4):
    res[i] = {}
    for cat in cats:
        res[i][cat] = 0
    res[i]['total'] = 0

print(res)
for row in data:
    idx = (int(row[4] / 10) - 2)
    if idx == -1: #if age is 18 or 19
        idx = 0
    if idx == 4: #if age is 60
        idx = 3
    print(idx)
    res[idx][row[10]] += row[11]
    res[idx]['total'] += row[11]
```

```python
import csv
import numpy as np

ind = ['18-29','30-39','40-49','50-60']

with open('/Users/franciscochoi/Desktop/infovis/tpVast/res/ej15.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    f.write("age" + ",")
    np.append(cats, 'amountReceived')
    writer.writerow(cats)
    it = 0
    for k in res:
        strRow = str(ind[it])
        for dk in res[k]:
            strRow += "," + str(res[k][dk])
        strRow += "\n"
        f.write(strRow)
        it += 1




print("Finish")
```
